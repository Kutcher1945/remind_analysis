```python
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Monto Google Drive

from google.colab import drive
drive.mount('/content/drive')
```

```python
data_dir = '/content/drive/MyDrive/Dataframes/best-alz-dataset/Combined Dataset/train'
```

```python
# Obtén la lista de clases
class_names = os.listdir(data_dir)
```

```python
# Inicializa un diccionario para almacenar el número de muestras por clase
class_counts = {}
```

```python
# Itera sobre cada clase y cuenta el número de muestras
for class_name in class_names:
    class_dir = os.path.join(data_dir, class_name)
    class_counts[class_name] = len(os.listdir(class_dir))
```

```python
# Calcula el número total de muestras
total_samples = sum(class_counts.values())
```

```python
# Imprime la información
print("Número de muestras en el conjunto de datos:", total_samples)
print("Número de clases en el conjunto de datos:", len(class_names))
print("Nombres de las clases:", class_names)
print("Número de muestras en cada clase:", class_counts)
```

```python
# Crea un gráfico de barras del número de muestras en cada clase
plt.bar(class_names, class_counts.values(), width=0.8, color='b')
plt.xlabel("Clase")
plt.ylabel("Número de muestras")
plt.title("Número de muestras en cada clase")
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()
```

```python
# Guarda el gráfico
plt.savefig('/content/drive/MyDrive/Dataframes/class_distribution.png')
```

```python
def count_nulls(data_dir):
    """Cuenta el número de píxeles nulos en un conjunto de datos de imágenes.

    Args:
        data_dir: La ruta al directorio que contiene las imágenes.

    Returns:
        Un diccionario que mapea los nombres de clase al número total de píxeles nulos en las imágenes de esa clase.
    """

    null_counts = {}  # Inicializa un diccionario para almacenar los conteos de nulos

    # Itera sobre cada clase y cada imagen dentro de la clase
    for class_name in os.listdir(data_dir):
        class_dir = os.path.join(data_dir, class_name)
        null_counts[class_name] = 0  # Inicializa el conteo de nulos para la clase actual
        for image_name in os.listdir(class_dir):
            image_path = os.path.join(class_dir, image_name)
            # Carga la imagen como un array NumPy
            image = np.array(Image.open(image_path))
            # Cuenta el número de píxeles nulos en la imagen y agrega al conteo de la clase
            null_counts[class_name] += np.count_nonzero(image == 0)
    return null_counts
```
